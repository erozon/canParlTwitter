```python
import tweepy
import os
import pandas as pd
```

```python
#Authentication, API Setup
PUB_KEY = os.environ.get("API_KEY")
PRIV_KEY = os.environ.get("API_SECRET_KEY")
auth = tweepy.AppAuthHandler(PUB_KEY, PRIV_KEY)
api = tweepy.API(auth)
```

```python
parlListID = 864088912087715840
```

```python
member_list = []
for mem in tweepy.Cursor(api.list_members, list_id=parlListID).items():
    member_list.append(mem._json)
```

```python
member_id_dict = {}
for mem in member_list:
    member_id_dict[mem['id_str']] = mem['name'] 
```

```python
member_id_dict
```

```python
def getMemberList():
    """
    A list of JSON objects, each containing all the information about an individual MP on Twitter. 
    """
    memberList = []
    for mem in tweepy.Cursor(api.list_members, list_id=parlListID).items():
        memberList.append(mem._json)
        
    return memberList

def getIdDict(memberList = None):
    """
    Returns a dictionary object linking MP Twitter ID (unique identifier) to MP Name. 
    """
    if memberList == None:
        memberList = getMemberList()
        
    memberIdDict = {}
    for mem in memberList:
        member_id_dict[mem['id_str']] = mem['name']
        
    return memberIdDict
```

```python
memberList = getMemberList()
```

```python
idDict = getIdDict(memberList)
```
